```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://www.caak.mn/")
```

```python
r.content
```

```python
soup = BeautifulSoup(r.content)
```

```python
print(soup.prettify())
```

```python
soup.find_all("a")
```

```python
for link in soup.find_all("a"):
    print(link)
    link.get("href")
```

```python
for link in soup.find_all("a"):
    print(link.text)
```

```python
for link in soup.find_all("a"):
    print(link.text, link.get("href"))
```

```python
for link in soup.find_all("a"):
    b = "<a href ='%s'>%s</a>"%(link.get("href"), link.text)
    print(b)
```
